#### Exercise

Write a function that takes a Series and returns the top 10% registers. In this case, earners. Test it with this Series:

```python
salaries = pd.Series([150000, 90000, 120000,30000,10000,5000,40000, 50000, 80000, 35000, 27000,14000, 28000, 22000,25000])
```

````python
salaries = pd.Series([150000, 90000, 120000,30000,10000,5000,40000, 50000, 80000, 35000, 27000,14000, 28000, 22000,25000])
salaries.rank(method="first", pct=True)
````

````python
def top10pct_earners(series):
    firstSalaries = salaries.rank(method="first", pct=True)
    top10 = firstSalaries[firstSalaries.values > 0.1]
    return series [top10.index]
````

````python
top10pct_earners(salaries)
````

````python
def top10pct_earners_3(series, percentile):
    topEarners = salaries.rank(method="first", pct=True) > percentile
    return series [topEarners].sort_values(ascending=False)
````

````python
top10pct_earners_3(salaries, .5)
````
